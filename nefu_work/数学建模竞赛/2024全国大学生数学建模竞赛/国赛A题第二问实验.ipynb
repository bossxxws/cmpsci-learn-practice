```python
import numpy as np

def calculate_vertices(hole1, hole2, vertex):
    # 小孔的坐标
    x1, y1 = hole1
    x2, y2 = hole2
    
    # 已知顶点的坐标
    x3, y3 = vertex
    
    # 长方形的长和宽
    length = 2.2
    width = 0.3
    
    # 小孔之间的距离
    hole_distance = 1.65
    
    # 小孔到宽边的距离
    hole_to_width_edge = 0.275
    
    # 小孔到长边的距离
    hole_to_length_edge = 0.15
    
    # 计算长方形的方向向量
    direction_vector = np.array([x2 - x1, y2 - y1])
    direction_vector = direction_vector / np.linalg.norm(direction_vector)
    
    # 计算长方形的四个顶点
    vertices = []
    vertices.append((x3, y3))
    vertices.append((x3 + length * direction_vector[0], y3 + length * direction_vector[1]))
    vertices.append(((x3 - width * direction_vector[1]), y3 + width * direction_vector[0]))
    vertices.append(((x3 + length * direction_vector[0] - width * direction_vector[1]), y3 + length * direction_vector[1] + width * direction_vector[0]))
    
    return vertices

# 示例输入
hole1 = (-0.15, -0.275)
hole2 = (-0.15, -1.925)
vertex = (0, 0)

# 计算其他三个顶点的坐标
vertices = calculate_vertices(hole1, hole2, vertex)
print("其他三个顶点的坐标为：", vertices)
```

```python
import numpy as np

def calculate_rectangle_vertices(hole1, hole2, vertex):
    x1, y1 = hole1
    x2, y2 = hole2
    x3, y3 = vertex

    # 计算小孔1和2相对于已知顶点的位置向量
    hole_vector1 = np.array([x1 - x3, y1 - y3])
    hole_vector2 = np.array([x2 - x3, y2 - y3])

    # 选择离已知顶点最近的孔
    hole_distance1 = np.linalg.norm(hole_vector1)
    hole_distance2 = np.linalg.norm(hole_vector2)

    if hole_distance1 < hole_distance2:
        parent_hole = hole1
        child_hole = hole2
    else:
        parent_hole = hole2
        child_hole = hole1

    x1, y1 = parent_hole
    x2, y2 = child_hole

    # 计算两个小孔连线的方向向量
    direction_vector = np.array([x2 - x1, y2 - y1])
    direction_vector = direction_vector / np.linalg.norm(direction_vector)  # 归一化，确保是单位向量

    # 计算垂直方向向量
    perpendicular_vector1 = np.array([-direction_vector[1], direction_vector[0]])
    perpendicular_vector2 = np.array([direction_vector[1], -direction_vector[0]])

    # 计算起始顶点到父小孔的单位向量
    start_to_parent = (hole_vector1 / np.linalg.norm(hole_vector1)) if hole_distance1 < hole_distance2 else (hole_vector2 / np.linalg.norm(hole_vector2))

    # 选择正确的垂直向量
    dot_product1 = np.dot(perpendicular_vector1, start_to_parent)
    dot_product2 = np.dot(perpendicular_vector2, start_to_parent)
    perpendicular_vector = perpendicular_vector1 if dot_product1 > dot_product2 else perpendicular_vector2

    # 长方形的长和宽
    length = 2.2
    width = 0.3

    # 计算另外三个顶点
    vertex2 = np.array([x3, y3]) + length * direction_vector
    vertex3 = np.array([x3, y3]) + width * perpendicular_vector
    vertex4 = vertex2 + width * perpendicular_vector

    return (x3, y3), tuple(vertex2), tuple(vertex3), tuple(vertex4)

# 示例
hole1 = (0.275, -0.15)
hole2 = (1.925, -0.15)
vertex = (2.2, 0)

# 计算四个顶点
vertices = calculate_rectangle_vertices(hole1, hole2, vertex)
print("长方形的四个顶点坐标为：", vertices)
```

```python
import matplotlib.pyplot as plt

# 定义矩形的四个顶点坐标
rectangle_points = [
    vertex,
    vertices[1],
    vertices[2],
    vertices[3]
]

# 定义额外的两个点的坐标
additional_points = [
    hole1,
    hole2
]

# 绘制矩形的四个顶点
plt.figure(figsize=(8, 6))
plt.plot(*zip(*rectangle_points), 'bo-', label='Rectangle Points')  # 'bo-' 表示蓝色圆点和实线

# 绘制额外的两个点
plt.plot(*zip(*additional_points), 'ro-', label='Additional Points')  # 'ro-' 表示红色圆点和实线

# 标记所有点
for i, (x, y) in enumerate(rectangle_points + additional_points, start=1):
    plt.text(x, y, f'P{i}', fontsize=12, ha='right')

# 设置图形属性
plt.grid(True)
plt.axis('equal')  # 确保x轴和y轴的比例相同
plt.title('Plot of Points')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend()

# 显示图形
plt.show()
```
